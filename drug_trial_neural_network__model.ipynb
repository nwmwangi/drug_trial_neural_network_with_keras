### A simple artificial neural network model using keras. The model is going to predict whether an individual experiences a side effect from a drug in a clinical trial. For this project, we are going to generate dummy data ourselves and then carry out the analysis.

#### import the needed libraries

```python
import numpy as np
import pandas as pd
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
"""Data containers (lists) for  our samples and labels"""
train_samples = []
train_labels = []
```

#### We are going to generate a dummy data that we are going to use in our analysis

###### The sample data assumes the following:
1. A clinical trial was carried out to test the effect a particular drug on a sample of participants aged beween 13 (thirteen) and 100 (a hundred years old)
2. The total number of participants was 2100 with half of them aged below 65 years and the other half aged above 65 years.
3. From the results, approximately 95% of the participants aged above 65 years registered side effects from the drug.
4. While 95% of the participants under the age of 65 years had no side effects at all.

```python
"""From the estimation above, we are going to generate a dummy dataset, the first for loop generates the 5% of particpants
from both categories of under 65 years and over 65 years."""

for i in range(50):
    """generate the approxmate 5% of participants under 65 years who exeprienced side effects (labeled 1)"""

    under_65 = randint(13,64)
    train_samples.append(under_65)
    train_labels.append(1)
    
    """generate the approxmate 5% of participants over 65 years who exeprienced no side effects (labeled 0)"""

    under_65 = randint(65,100)
    train_samples.append(under_65)
    train_labels.append(0)
    
print(len(train_samples))
print(len(train_labels))

for i in range(1000):
    """generate the approxmate 95% of participants under 65 years who exeprienced no side effects (labeled 0)"""

    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    """generate the approxmate 95% of participants over 65 years who exeprienced side effects (labeled 1)"""

    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
print(len(train_samples))
print(len(train_labels))
```

```python
"""Now, we transform the generated datasets into a numpy array for our processing """

train_samples = np.array(train_samples)
train_labels = np.array(train_labels)
"""Shuffle the datasets against each other to get rid of having ordered data from the generated list"""
train_labels, train_samples = shuffle(train_labels, train_samples)
```

### With the datasets in a numpy arrary, we scale the data to create the features or ranges 0 to 1.  We then feed reshaped dataset into the fit_transform() function as a 2 dimensional data.

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

### Use sequential model form the  tf.keras api to create artificial neural network

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

###### Build a simple sequential model: this is basically a linear stack of layers

```python
"""Initialize the model as an instance of sequential class, to create the model
    With a the first layers initialized with 16 nodes and the output layer with 2 nodes
    implying the two possible classes of the output"""

ann_model = Sequential([
    Dense(units=18, input_shape=(1,), activation='relu'),
    Dense(units=36, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
ann_model.summary()
```

### Prepare the model for training by getting the parameters in order.

```python
ann_model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### We train the model by calling the fit() function on:

    x=> input data, 
    y=> the labels or target data,
    batch_size=> number of samples included in one batch for each iteration of neural network processing,
    epochs=>number of times the model is going to train before it completes the training process,
    shuffle=>remove any order on the datasets before training begins,
    verbose=> enable output messages 
    validation_split=> splits a portion of the training data to create a validation dataset during train.This happens before the training set is shuffled.

```python
ann_model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=10, shuffle=True, verbose=2)
```

### Create a test dataset in order to use it inference how well the model would perform on a new dataset other than the validation data. Let's generate the tests datasets, in the same way, we generated the training dataset.

```python
test_samples = []
test_labels = []
```

```python
"""From the estimation above, we are going to generate a dummy dataset, the first for loop generates the 5% of particpants
from both categories of under 65 years and over 65 years."""

for i in range(50):
    """generate the approxmate 5% of participants under 65 years who exeprienced side effects (labeled 1)"""

    under_65 = randint(13,64)
    test_samples.append(under_65)
    test_labels.append(1)
    
    """generate the approxmate 5% of participants over 65 years who exeprienced no side effects (labeled 0)"""

    under_65 = randint(65,100)
    test_samples.append(under_65)
    test_labels.append(0)
    
print(len(test_samples))
print(len(test_labels))

for i in range(1000):
    """generate the approxmate 95% of participants under 65 years who exeprienced no side effects (labeled 0)"""

    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    """generate the approxmate 95% of participants over 65 years who exeprienced side effects (labeled 1)"""

    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
print(len(test_samples))
print(len(test_labels))
```

```python
test_samples = np.array(test_samples)
test_labels = np.array(test_labels)
test_labels, test_samples = shuffle(test_labels, test_samples)
```

```python

scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

## Now, we use the created model to predict on the test datasets by calling the predict() function on our trained model.

```python
predictions = ann_model.predict(x=scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
    print(i)
```

#### To get the most probable predication from the test dataset, we can take the index of the sample with the highest probability as follows:

```python
rounded_predictions = np.argmax(predictions, axis=-1)
```

```python
predicted_results = []
for i in rounded_predictions:
    predicted_results.append(i)
print(predicted_results)
```

###### From the predictions above, a 1 shows a higher probability of experiencing side effects, while a 0(zero) shows probability of not exeperiencing side effects.

### To assess how well the model performed on the test data, we can visualize the model's prediction accuracy through a confusion matrix. 

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix, accuracy_score
import itertools
import matplotlib.pyplot as plt
```

#### We pass the test labels as the truth/actual values and the predictions to the sklearn confusion matrix function.

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python
"""Function source: sklearn website"""

def plot_confusion_matrix(cm, classes, normalize=False, title="Confusion matrix", cmap=plt.cm.Reds):
    """
    This fuctions prints and plots the Confusion matrix.
    Normalization can be applied be setting `normalize=True`
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normailzed Confusion Matrix")
    else:
        print("Confusion matrix, without normalization")
    print(cm)
    
    thresh = cm.max() /2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j , i, cm[i, j], horizontalalignment='center', color='white' if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```

```python
cm_plot_labels = ["no_side_effects", "had_side_effects"]
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title="Confusion Matrix")

print(f"The accuracy score of the model: {accuracy_score(test_labels, rounded_predictions)}")
```

###### From the confusion matrix above, reading the diagonal from the top_left -- bottom_right, these are the true predictions  which by simple calculations adds up to approxiamate 87.1%  accuracy, which is not too off the 95% mark of our training data set.
